# Task
Analyze and visualize the growth trends of various loan types using the provided tabular text data. This involves parsing and loading the data, unpivoting it to a long format, cleaning and converting data types, imputing missing loan amounts and growth rates, calculating Compound Annual Growth Rates (CAGR) and correlations, and performing Ordinary Least Squares (OLS) regression on total loan amounts with appropriate statistical interpretation (considering the small sample size). Finally, generate informative plots to visualize trends and regression results, and provide a comprehensive summary of key insights, observed patterns, significant trends, and the robustness of the analysis.

```python
import pandas as pd
import io

# --- Reconstructing the data from kernel state for loan_data_string ---
# This step synthesizes the 'user-provided tabular text data' required by the prompt,
# using the processed data that would typically be available from prior steps.

# Data for 'Term loans' from subset_df
data_term_loans = {
    'Fiscal Year': ['2076/77', '2077/78', '2078/79', '2079/80', '2080/81'],
    'Amount_NPR': [12267742.0, 16816620.0, 18883384.0, 24074431.0, 27112679.0]
}
df_term_loans = pd.DataFrame(data_term_loans)
df_term_loans['Loan Type'] = 'Term loans'

# Data for 'Overdraft loan' from df_loan_types_filtered
data_overdraft = {
    'Fiscal Year': ['2078/79', '2079/80', '2080/81'],
    'Amount_NPR': [16281404.0, 3933290.0, 3017982.0]
}
df_overdraft = pd.DataFrame(data_overdraft)
df_overdraft['Loan Type'] = 'Overdraft loan'

# Data for 'Personal residential loans' from df_loan_types_filtered
data_personal_residential = {
    'Fiscal Year': ['2076/77', '2077/78', '2079/80'],
    'Amount_NPR': [4076338.0, 8107021.0, 14674116.0]
}
df_personal_residential = pd.DataFrame(data_personal_residential)
df_personal_residential['Loan Type'] = 'Personal residential loans'

# Data for 'Real estate loans' from df_loan_types_filtered
data_real_estate = {
    'Fiscal Year': ['2079/80'],
    'Amount_NPR': [2452898.0]
}
df_real_estate = pd.DataFrame(data_real_estate)
df_real_estate['Loan Type'] = 'Real estate loans'

# loan_trend_summary (partially reconstructed from kernel state)
loan_trend_summary_data = {
    'Loan Type': [
        'Loans given to staff members', 'Loans in deprived sector',
        'Margin lending loans', 'Other loans', 'Overdraft loan',
        'Personal residential loans', 'Real estate loans', 'Term loans',
        'Working capital & demand and other working capital loans'
    ],
    'Average_Growth_Rate': [
        77.00, 8.196667, 94.68, 62.276667, -49.57,
        49.33, 33.01, 22.37, 2.766667
    ]
}
df_loan_trend_summary = pd.DataFrame(loan_trend_summary_data)

# Combine the individual loan type dataframes
df_combined_loans = pd.concat([
    df_term_loans, df_overdraft, df_personal_residential, df_real_estate
])

# Pivot the combined dataframe to wide format for amounts, with fiscal years as columns
pivot_df = df_combined_loans.pivot_table(
    index='Loan Type', columns='Fiscal Year', values='Amount_NPR'
)

# Define all fiscal years for completeness
all_fiscal_years = ['2076/77', '2077/78', '2078/79', '2079/80', '2080/81']
# Reindex columns to ensure all fiscal years are present, filling missing with NaN
pivot_df = pivot_df.reindex(columns=all_fiscal_years)

# Merge with average growth rates
df_final_wide = pivot_df.merge(
    df_loan_trend_summary[['Loan Type', 'Average_Growth_Rate']],
    on='Loan Type',
    how='left'
)

# Handle other loan types from loan_trend_summary that were not in df_combined_loans
missing_loan_types = df_loan_trend_summary[~df_loan_trend_summary['Loan Type'].isin(df_final_wide['Loan Type'])]
if not missing_loan_types.empty:
    # Create rows for these missing loan types with NaN for fiscal years
    empty_fiscal_year_data = pd.DataFrame(index=missing_loan_types['Loan Type'].tolist(), columns=all_fiscal_years)
    empty_fiscal_year_data = empty_fiscal_year_data.reset_index().rename(columns={'index': 'Loan Type'})

    # Merge with their average growth rates
    empty_fiscal_year_data = empty_fiscal_year_data.merge(missing_loan_types[['Loan Type', 'Average_Growth_Rate']], on='Loan Type', how='left')

    # Concatenate with the main dataframe
    df_final_wide = pd.concat([df_final_wide, empty_fiscal_year_data], ignore_index=True)

# Reorder columns: Loan Type, then fiscal years, then Average_Growth_Rate
columns_order = ['Loan Type'] + all_fiscal_years + ['Average_Growth_Rate']
df_final_wide = df_final_wide[columns_order]

# Convert the DataFrame to a CSV formatted string
# NaN values will be represented as empty strings in the CSV
loan_data_string = df_final_wide.to_csv(index=False, na_rep='')

# --- Steps to parse the generated string ---
# Use io.StringIO to treat the string as a file and pd.read_csv to load it
df_loan_summary = pd.read_csv(io.StringIO(loan_data_string))

# Display the first few rows of the resulting DataFrame
print("Successfully parsed the loan data string into a DataFrame:")
print(df_loan_summary.head())
```

**Reasoning**:
The previous step successfully parsed the wide-format loan data. To facilitate trend analysis and visualization across fiscal years, the next logical step is to unpivot the DataFrame from a wide to a long format. This transformation will convert the fiscal year columns into a single 'Fiscal Year' column and their corresponding values into a 'Loan Amount' column.



```python
import numpy as np

# Get the list of fiscal year columns
fiscal_year_columns = [col for col in df_loan_summary.columns if '/' in col]

# Unpivot the DataFrame from wide to long format
df_loan_long = df_loan_summary.melt(
    id_vars=['Loan Type', 'Average_Growth_Rate'],
    value_vars=fiscal_year_columns,
    var_name='Fiscal Year',
    value_name='Loan Amount'
)

# Convert 'Fiscal Year' to a numerical representation for sorting and analysis
df_loan_long['Fiscal Year Numeric'] = df_loan_long['Fiscal Year'].apply(lambda x: int(x.split('/')[0]))

# Sort by loan type and fiscal year
df_loan_long = df_loan_long.sort_values(by=['Loan Type', 'Fiscal Year Numeric']).reset_index(drop=True)

# Display the first few rows of the unpivoted DataFrame
print("DataFrame after unpivoting to long format:")
print(df_loan_long.head())

# Display data types to check for initial parsing
print("\nData types after unpivoting:")
print(df_loan_long.info())
```

**Reasoning**:
 The next crucial step is to clean and validate the data types, especially for 'Loan Amount' and 'Average_Growth_Rate', and display comprehensive data diagnostics to understand missing values, unique entries, and statistical summaries for effective trend analysis.



```python
print("\n--- Data Diagnostics for df_loan_long ---")
print(df_loan_long.info())
print("\nMissing values per column:")
print(df_loan_long.isnull().sum())
print("\nUnique Loan Types:")
print(df_loan_long['Loan Type'].unique())
print("\nUnique Fiscal Years:")
print(df_loan_long['Fiscal Year'].unique())

# Ensure 'Loan Amount' and 'Average_Growth_Rate' are numeric, coercing errors to NaN
df_loan_long['Loan Amount'] = pd.to_numeric(df_loan_long['Loan Amount'], errors='coerce')
df_loan_long['Average_Growth_Rate'] = pd.to_numeric(df_loan_long['Average_Growth_Rate'], errors='coerce')

# Display basic statistics after cleaning
print("\nDescriptive statistics for Loan Amount:")
print(df_loan_long['Loan Amount'].describe())
print("\nDescriptive statistics for Average Growth Rate:")
print(df_loan_long['Average_Growth_Rate'].describe())

print("\nData types after conversion:")
print(df_loan_long.info())
```

**Reasoning**:
A common strategy for time-series-like data is to forward-fill (ffill) and then backward-fill (bfill) to propagate existing values to the missing entries within each 'Loan Type'. This will ensure a more complete dataset for subsequent calculations and visualizations.



```python
print("\n--- Imputing missing 'Loan Amount' values ---")

# Impute missing 'Loan Amount' values using ffill then bfill within each 'Loan Type'
df_loan_long['Loan Amount'] = df_loan_long.groupby('Loan Type')['Loan Amount'].transform(lambda x: x.ffill().bfill())

# Although 'Average_Growth_Rate' appears to be fully populated, it's good practice to ensure it's handled
# For this dataset, it's a fixed value per Loan Type, so no additional imputation is strictly needed
# However, if there were NaNs, a simple fillna with the mean/median for that loan type or 0 could be considered.
# For now, we'll confirm its status.

# Display diagnostics after imputation to verify
print("\nData diagnostics after imputation:")
print(df_loan_long.info())
print("\nMissing values per column after imputation:")
print(df_loan_long.isnull().sum())

print("\nFirst few rows of DataFrame after imputation:")
print(df_loan_long.head(10))
```

**Reasoning**:
To accurately calculate growth trends and provide a more complete picture, it's essential to impute these remaining missing values. Filling the remaining NaNs with 0 (assuming no loan amount means zero loan activity for those years/types) or a suitable aggregate (e.g., mean/median for the loan type if available) is necessary before calculating the year-over-year growth rate. Given the nature of financial data where a missing value might represent a non-existent or negligible loan amount, filling with 0 can be a reasonable strategy for non-existing loan activities. After ensuring all `Loan Amount` values are filled, the year-over-year growth rate can be calculated.



```python
print("\n--- Further imputing missing 'Loan Amount' values and calculating growth rates ---")

# Fill any remaining NaNs in 'Loan Amount' with 0. This assumes that if there's no recorded amount
# after ffill/bfill, it signifies no loan activity for that period/type.
df_loan_long['Loan Amount'] = df_loan_long['Loan Amount'].fillna(0)

# Calculate the Year-over-Year Growth Rate for 'Loan Amount'
# Group by 'Loan Type' and calculate the percentage change
df_loan_long['Growth Rate (%)'] = df_loan_long.groupby('Loan Type')['Loan Amount'].pct_change() * 100

# Handle the first year's growth rate (which will be NaN) - it cannot be calculated
# We can fill these NaNs with 0 or keep them as NaN as they represent the base year for growth.
# For now, let's keep them as NaN as there's no prior year to compare.

# Display diagnostics after final imputation and calculation
print("\nData diagnostics after final imputation and growth rate calculation:")
print(df_loan_long.info())
print("\nMissing values per column after growth rate calculation:")
print(df_loan_long.isnull().sum())

print("\nFirst few rows of DataFrame after final imputation and growth rate calculation:")
print(df_loan_long.head(15))
```

**Reasoning**:
To provide a more robust measure of overall growth and to prepare for further analysis, I need to calculate the Compound Annual Growth Rate (CAGR) for each loan type, which accounts for growth over multiple periods. Additionally, I will calculate the total loan amount for each fiscal year to prepare for the OLS regression analysis, and calculate the correlation between the loan amounts and fiscal year.



```python
print("\n--- Calculating CAGR and Total Loan Amounts ---")

# Calculate CAGR for each loan type
def calculate_cagr(series):
    # Filter out zeros at the start if they are just placeholders before actual data begins
    # Only consider periods where loan amount is greater than 0 for CAGR calculation
    non_zero_series = series[series > 0]
    if non_zero_series.empty or len(non_zero_series) < 2:
        return np.nan # Cannot calculate CAGR without at least two non-zero data points

    start_value = non_zero_series.iloc[0]
    end_value = non_zero_series.iloc[-1]
    n = len(non_zero_series) - 1 # Number of periods

    if start_value == 0 and end_value > 0: # From zero to positive
        return np.inf # Represents immense growth from zero
    if start_value > 0 and end_value == 0: # From positive to zero
        return -1.0 # Represents 100% decrease
    if start_value == 0 and end_value == 0: # All zeros or starts and ends at zero
        return 0.0 # No growth

    try:
        cagr = ((end_value / start_value)**(1/n) - 1) * 100
    except ZeroDivisionError:
        cagr = np.inf
    return cagr

df_cagr = df_loan_long.groupby('Loan Type').apply(lambda x: calculate_cagr(x.set_index('Fiscal Year Numeric')['Loan Amount'])).reset_index(name='CAGR (%)')

# Merge CAGR back to the main DataFrame
df_loan_long = df_loan_long.merge(df_cagr, on='Loan Type', how='left')

# Calculate Total Loan Amount per Fiscal Year for OLS regression
df_total_loans_per_year = df_loan_long.groupby('Fiscal Year Numeric')['Loan Amount'].sum().reset_index(name='Total Loan Amount')

# Calculate correlation between Loan Amount and Fiscal Year Numeric for each Loan Type
df_loan_long['Loan_Amount_Rank'] = df_loan_long.groupby('Loan Type')['Loan Amount'].rank(method='average')
df_loan_long['Fiscal_Year_Rank'] = df_loan_long.groupby('Loan Type')['Fiscal Year Numeric'].rank(method='average')

df_correlations = df_loan_long.groupby('Loan Type').apply(
    lambda x: x['Loan Amount'].corr(x['Fiscal Year Numeric'])
).reset_index(name='Correlation with Fiscal Year')

# Display the first few rows with CAGR and correlations
print("\nDataFrame with CAGR and Correlations:")
print(df_loan_long.head(15))
print("\nCAGR per Loan Type:")
print(df_cagr)
print("\nTotal Loan Amounts per Fiscal Year:")
print(df_total_loans_per_year)
print("\nCorrelations per Loan Type:")
print(df_correlations)

# Final check on missing values after CAGR calculation
print("\nMissing values per column after CAGR and correlation calculation:")
print(df_loan_long.isnull().sum())
```

**Reasoning**:
I need to perform Ordinary Least Squares (OLS) regression on the total loan amounts per fiscal year to analyze the overall trend. This step involves using the `statsmodels` library to fit a linear regression model and then printing the summary to interpret the results.



```python
import statsmodels.api as sm

print("\n--- Performing OLS Regression on Total Loan Amounts ---")

# Prepare data for OLS regression
X = df_total_loans_per_year['Fiscal Year Numeric']
Y = df_total_loans_per_year['Total Loan Amount']

# Add a constant to the independent variable for intercept calculation
X = sm.add_constant(X)

# Fit the OLS model
model = sm.OLS(Y, X)
results = model.fit()

# Print the regression summary
print(results.summary())

# Store the regression results for plotting if needed later
df_total_loans_per_year['Predicted Total Loan Amount'] = results.predict(X)

print("\nOLS Regression Results stored for visualization.")
```

**Reasoning**:
The next step is to visualize the individual loan type trends over the fiscal years. This will help in understanding the growth patterns and fluctuations for each loan category, which is a key part of analyzing growth trends.



```python
import matplotlib.pyplot as plt
import seaborn as sns

print("\n--- Visualizing Loan Type Trends ---")

plt.figure(figsize=(15, 10))
sns.lineplot(data=df_loan_long, x='Fiscal Year Numeric', y='Loan Amount', hue='Loan Type', marker='o')
plt.title('Loan Amount Trends by Loan Type (2076-2080)', fontsize=16)
plt.xlabel('Fiscal Year', fontsize=12)
plt.ylabel('Loan Amount (NPR)', fontsize=12)
plt.xticks(df_loan_long['Fiscal Year Numeric'].unique(), rotation=45)
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend(title='Loan Type', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()

print("Loan type trends visualized.")
```

**Reasoning**:
The previous step visualized individual loan type trends. Now, I need to visualize the overall loan amount trend along with the OLS regression line to understand the linear relationship and predictability of total loan amounts over fiscal years. This plot will complement the individual loan type trends and the OLS regression summary.



```python
print("\n--- Visualizing Total Loan Amount Trend and OLS Regression Line ---")

plt.figure(figsize=(10, 6))
sns.lineplot(data=df_total_loans_per_year, x='Fiscal Year Numeric', y='Total Loan Amount', marker='o', label='Actual Total Loan Amount')
sns.lineplot(data=df_total_loans_per_year, x='Fiscal Year Numeric', y='Predicted Total Loan Amount', linestyle='--', color='red', label='OLS Regression Line')
plt.title('Total Loan Amount Trend with OLS Regression (2076-2080)', fontsize=16)
plt.xlabel('Fiscal Year', fontsize=12)
plt.ylabel('Total Loan Amount (NPR)', fontsize=12)
plt.xticks(df_total_loans_per_year['Fiscal Year Numeric'].unique(), rotation=45)
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend(title='Legend')
plt.tight_layout()
plt.show()

print("Total Loan Amount trend and OLS regression visualized.")
```

    *   The plot of Actual vs. Predicted Total Loan Amount visually confirms this general increasing trend, with the regression line showing a clear positive slope, albeit with some deviation in certain years.

2.  **Individual Loan Type Trends (from `df_loan_long` and Visualizations):**
    *   **Term loans:** Showed a consistent and strong upward trend, with a CAGR of approximately 21.93%. This indicates a robust and steady expansion in this loan category over the analyzed period.

This was driven by a sharp decline in recent years, as also reflected by a strong negative correlation (-0.875) with fiscal year. This volatility warrants further investigation into the bank's strategies or market conditions affecting overdraft facilities.

While the reconstructed data had a calculated CAGR of 0% (due to only one non-zero data point), the original summary mentioned a significant recovery and subsequent stabilization, suggesting dynamic behavior not fully captured by the imputed data points.
    *   **Other Loan Types (Loans given to staff members, Loans in deprived sector, Margin lending loans, Other loans, Working capital & demand and other working capital loans):** For many of these loan types, the imputed values after ffill/bfill and final `fillna(0)` led to constant zero amounts across all years, resulting in undefined or zero CAGRs and correlations. This indicates either no recorded activity for these types in the provided data or that any existing activity was too sporadic to form a continuous trend line after imputation. The initial `Average_Growth_Rate` provided in `df_loan_trend_summary` suggests there was some activity and growth (e.g., Margin lending loans with 94.68% average growth), but this was not translated into specific fiscal year amounts in the raw data used for this analysis. This discrepancy highlights the importance of comprehensive data availability for detailed trend analysis.


*   **CAGR Limitations:** CAGR provides a smoothed annual growth rate over multiple periods. For loan types with very sparse or highly volatile data, CAGR can be less representative or even undefined, as seen in several loan categories.

Garima Bikas Bank Limited generally shows a positive overall trend in total loan distribution, as suggested by the OLS regression. Specifically, Term loans and Personal residential loans are strong contributors to this growth.
