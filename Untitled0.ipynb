```python
import pandas as pd

# Update this path to match your exact file location in Colab
file_path = '/content/drive/MyDrive/Colab Notebooks/Garima Bikas Bank Limited - Loan Distribution Data - Sheet1.csv'  # â† adjust if in Drive or renamed

try:
    # Load with flexible options to handle common export issues
    # Added skiprows=4 to skip initial descriptive rows
    df = pd.read_csv(file_path, encoding='utf-8-sig', thousands=',', skipinitialspace=True, skiprows=4)

    # Set the first row as column headers and then drop it
    df.columns = df.iloc[0] # Use the first row as column names
    df = df[1:].reset_index(drop=True) # Drop the first row and reset index

    print("=== CSV Diagnostic Report ===")
    print("Shape (rows, columns):", df.shape)
    print("\nColumn names:\n", df.columns.tolist())
    print("\nFirst 3 rows (head):\n", df.head(3).to_string(index=False))
    print("\nLast 2 rows (tail):\n", df.tail(2).to_string(index=False))
    print("\nData types:\n", df.dtypes)
    print("\nMissing values per column:\n", df.isnull().sum())

    # Identify numeric loan columns automatically (exclude Fiscal_Year)
    possible_loan_cols = [col for col in df.columns if col != 'Fiscal Year' and df[col].dtype in ['float64', 'int64']]
    print("\nDetected numeric loan columns:", possible_loan_cols)

    # Basic summary stats on loan amounts
    if possible_loan_cols:
        print("\nSummary statistics (numeric columns):\n", df[possible_loan_cols].describe().round(0))

    # If 'Total_Loan' exists, validate summation
    if 'Total_Loan' in df.columns and len(possible_loan_cols) > 1:
        sum_cols = [c for c in possible_loan_cols if c != 'Total_Loan']
        df['Sum_Check'] = df[sum_cols].sum(axis=1, skipna=True)
        df['Discrepancy'] = df['Total_Loan'] - df['Sum_Check']
        print("\nTotal verification (Discrepancy should be ~0):\n",
              df[['Fiscal_Year', 'Total_Loan', 'Sum_Check', 'Discrepancy']].round(0).to_string(index=False))

except FileNotFoundError:
    print("File not found. Confirm name/path. Use !ls to list files in current directory.")
except Exception as e:
    print("Loading error:", str(e))
    print("Try: pd.read_csv(..., encoding='latin1') or open in text editor to inspect delimiters/quotes.")
```

```python
columns_to_clean = [
    "Total Deposit (Rs. in '000)",
    "Total Loan & Advances (Rs. in '000)"
]

for col in columns_to_clean:
    if col in df.columns:
        # Ensure the column is treated as string before applying .str.replace()
        df[col] = df[col].astype(str).str.replace(',', '', regex=False)

# Clean 'CD Ratio (%)' column
if 'CD Ratio (%)' in df.columns:
    # Ensure the column is treated as string before applying .str.replace()
    df['CD Ratio (%)'] = df['CD Ratio (%)'].astype(str).str.replace('%', '', regex=False)

# Convert the cleaned columns to numeric, coercing errors to NaN
all_numeric_cols = columns_to_clean + ['CD Ratio (%)']
for col in all_numeric_cols:
    if col in df.columns:
        df[col] = pd.to_numeric(df[col], errors='coerce')

print("=== After cleaning and type conversion ===")
print("Data types after conversion:\n", df.dtypes)
print("\nFirst 5 rows after conversion:\n", df.head().to_string())
print("\nMissing values per column after conversion:\n", df.isnull().sum())
```

```python
df.dropna(subset=['Fiscal Year'], inplace=True)

# Clean the 'Fiscal Year' column
# Extract the first four characters (starting year) and convert to integer
df['Fiscal Year'] = df['Fiscal Year'].astype(str).str[:4]
df['Fiscal Year'] = pd.to_numeric(df['Fiscal Year'], errors='coerce')

# Drop rows where 'Fiscal Year' became NaN after conversion
df.dropna(subset=['Fiscal Year'], inplace=True)

print("=== After cleaning 'Fiscal Year' and removing non-data rows ===")
print("Data types after conversion:\n", df.dtypes)
print("\nFirst 5 rows after conversion:\n", df.head().to_string())
print("\nMissing values per column after conversion:\n", df.isnull().sum())
```

```python
df.dropna(subset=all_numeric_cols, inplace=True)

# Convert 'Fiscal Year' to integer type since it's now clean and ready for analysis
df['Fiscal Year'] = df['Fiscal Year'].astype(int)

print("=== After dropping rows with missing numerical data ===")
print("Shape (rows, columns):", df.shape)
print("Data types after final cleaning:", df.dtypes)
print("\nFirst 5 rows of the cleaned DataFrame:\n", df.head().to_string())
print("\nMissing values per column after final cleaning:\n", df.isnull().sum())
```

```python
print("=== Updated CSV Diagnostic Report ===")
print("Shape (rows, columns):", df.shape)
print("\nColumn names:\n", df.columns.tolist())
print("\nFirst 3 rows (head):\n", df.head(3).to_string(index=False))
print("\nLast 2 rows (tail):\n", df.tail(2).to_string(index=False))
print("\nData types:\n", df.dtypes)
print("\nMissing values per column:\n", df.isnull().sum())

# Identify numeric loan columns automatically (exclude Fiscal_Year)
possible_loan_cols = [col for col in df.columns if col != 'Fiscal Year' and df[col].dtype in ['float64', 'int64']]
print("\nDetected numeric loan columns:", possible_loan_cols)

# Basic summary stats on loan amounts
if possible_loan_cols:
    print("\nSummary statistics (numeric columns):\n", df[possible_loan_cols].describe().round(0))
```

```python
df['Deposit Growth (%)'] = df['Total Deposit (Rs. in \'000)'].pct_change() * 100
df['Loan Growth (%)'] = df['Total Loan & Advances (Rs. in \'000)'].pct_change() * 100

average_cd_ratio = df['CD Ratio (%)'].mean()

print("=== DataFrame with Growth Rates ===")
print(df.to_string())
print(f"\nAverage CD Ratio (%): {average_cd_ratio:.2f}")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(15, 5))

# Plot Total Deposit
plt.subplot(1, 3, 1) # 1 row, 3 columns, 1st plot
sns.lineplot(x='Fiscal Year', y="Total Deposit (Rs. in '000)", data=df)
plt.title('Total Deposit Over Fiscal Years')
plt.xlabel('Fiscal Year')
plt.ylabel("Total Deposit (Rs. in '000)")

# Plot Total Loan & Advances
plt.subplot(1, 3, 2) # 1 row, 3 columns, 2nd plot
sns.lineplot(x='Fiscal Year', y="Total Loan & Advances (Rs. in '000)", data=df)
plt.title('Total Loan & Advances Over Fiscal Years')
plt.xlabel('Fiscal Year')
plt.ylabel("Total Loan & Advances (Rs. in '000)")

# Plot CD Ratio
plt.subplot(1, 3, 3) # 1 row, 3 columns, 3rd plot
sns.lineplot(x='Fiscal Year', y='CD Ratio (%)', data=df)
plt.title('CD Ratio Over Fiscal Years')
plt.xlabel('Fiscal Year')
plt.ylabel('CD Ratio (%)')

plt.tight_layout()
plt.show()
```
